```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head()
```

```python
df['keywords'] = df['keywords'].apply(convert)
df.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
df['cast'] = df['cast'].apply(convert)
```

```python
df.head()
```

```python
df['cast'] = df['cast'].apply(lambda x:x[0:3])
```

```python
def get_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
df['crew'] = df['crew'].apply(get_director)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df1 = df.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
df1['tags'] = df1['tags'].apply(lambda x: " ".join(x))
```

```python
df1['tags'] = df1['tags'].apply( lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
df1['tags'] = df1['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(df1['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
df1[df1['original_title'] == 'Avatar'].index[0]
```

```python
def recommend(movie):
    if movie in df1['original_title'].values:
        index = df1[df1['original_title'] == movie].index[0]
        distances = sorted(enumerate(similarity[index]), reverse=True, key=lambda x: x[1])
        recommended_movies = [df1.iloc[i[0]].original_title for i in distances[1:6]]
        return recommended_movies
    else:
        return []

while True:
    Recently_watched_movie = input("Enter a movie name (or type 'exit' to quit): ")

    if Recently_watched_movie.lower() == 'exit':
        break

    recommended_movies = recommend(Recently_watched_movie)

    if recommended_movies:
        print("Recommended movies:")
        for movie in recommended_movies:
            print(movie)
    else:
        print("Movie not found in the database.")
```
